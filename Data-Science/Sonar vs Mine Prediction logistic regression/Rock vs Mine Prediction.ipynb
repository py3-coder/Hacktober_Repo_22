# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection & Data processing

```python
#Loading the dataset to a pandas Dataframe
sonar_data = pd.read_csv('SonarData.csv',header=None)
```

```python
sonar_data.head()
```

```python
#Number of rows and columns
sonar_data.shape
```

```python
#decription --> statistical measures of the data
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
# the no of mines and rocks examples datasets is almost equal hence we can get a good accuracy or a good model
#more the data , more better the model is
```

```python
sonar_data.groupby(60).mean()
```

```python
# separating labels and data
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

# Training and Test data 

```python
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1, stratify = Y, random_state = 1)
# test_size denotes that 0.1 i.e. 10% of the data is there in test data
#stratify = this divides our data evenly , means in test data we have around equal no of M and R 
# random state = data is splitted in a specific way we can use random_state = 2 as well
```

```python
print(X.shape,X_train.shape,X_test.shape,Y_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data 
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
#accuracy of the model on training data > accuracy on test data
# accuracy greater than 70% is good
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
# accuracy on test data 
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(testing_data_accuracy)
```
